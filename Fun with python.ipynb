# Forward difference

```python
def forward_difference(f,x,h=(1e-2)):
    return (f(x+h)-f(x))/h

def analytical_derivative(x=1, a=0.1, b=10):
    return np.exp(-a*(x**2))*((b*np.cos(b*x))-(2*a*x*(np.sin(b*x))))

print(forward_difference(f,x=1, h=(1e-2)))
print(analytical_derivative(x=1))
```

# Central difference

```python
def central_difference(f,x=1, h=(1e-2)):
    return 0.5*(f(x+h)-f(x-h))/(h)

print(central_difference(f,x=1, h=(1e-2)))
```

```python
s=np.logspace(-16,0,10)
fd_error=np.abs(analytical_derivative()-forward_difference(f,x=1,h=s))
cd_error=np.abs(analytical_derivative()-central_difference(f,x=1,h=s))

plt.plot(s,fd_error,'-^',label='forward difference')
plt.plot(s,cd_error,'-*', label='central difference')
plt.xscale('log')
plt.yscale('log')
plt.legend()
plt.grid()
```

# The numerical error has a minimum at 2*10e-8 and 1.56e-4 for the forward and backward difference respectively, it does not continue to decrease beyond this point for both errors rather it increase. The explanation for this behavior is two competing effects: truncation errors, roundoff errors. 
# However, the roundoff error is random in nature and will change from machine to machine, and each time we run the program. Note that the roundoff error increases when  the numerical error decreases, and the truncation error decreases when the numerical error decreases.

# Taylors formula only takes into account the truncation error and not the roundoff error. according to taylors formula the error are never meant to decrease but this is not the case due to the roundoff errors

```python
surface_elevation_values = bed1d['surface'].values
thickness_values= bed1d['surface'].values
bedrock_values=bed1d['bed'].values
hw=-(surface_elevation_values-thickness_values)

density_water= 997 #kg/m3
density_ice= 917 #kg/m3
density_ice_water= 980 #kg/m3
freeboard_level=hw*((density_water/density_ice)-1)
x1=np.linspace(0,250,len(bed1d['surface']))

h_ice_above = surface_elevation_values-freeboard_level
volume_above_freeboard = (h_ice_above*(.1e06)) 
total_volume_ice=np.nansum(volume_above_freeboard)
total_volume_ice_water=(density_ice/density_ice_water)*total_volume_ice
total_volume_water=(density_ice_water/density_water)*total_volume_ice_water
net_volume=total_volume_ice_water-total_volume_water
slr_graph=(thickness_values+(-bedrock_values))*((density_ice_water/density_ice)-1) #(0.0687)
sea_water_area = (361*10e6)
sea_level_rise = net_volume*1000/sea_water_area
print("sea level rise in meters:", sea_level_rise)
plt.plot(x1,bedrock_values, label="bed rock", color='k')
plt.plot(x1,thickness_values,label="thickness",color='b')
plt.plot(x1,freeboard_level, label="freeboard", color='g')
plt.plot(x1,slr_graph, label="ice above leads to slr", color='r')
plt.legend()
plt.title('Doomsday glacier')
plt.xlabel(' distance along track[km]')
plt.ylabel('height above sea level')

```
